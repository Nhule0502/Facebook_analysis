```python
import json
# Open and read the JSON file
import pandas as pd
df = pd.read_json('/Users/mac/Documents/Facebook_analysis/facebook-nhukyu3-29_07_2024-vtLS7dct/your_facebook_activity/posts/your_posts__check_ins__photos_and_videos_1.json', encoding='utf-8')

# Extract the year 
df['year'] = df['timestamp'].dt.to_period('Y')

# Group by year and count the occurrences
year_counts = df.groupby('year').size().reset_index(name='photo and video count')

# Print the result
print(year_counts)

import matplotlib.pyplot as plt
import numpy as np



# Extract the x (year_month) and height (count) values for the bar plot
x = year_counts['year'].astype(str) 
height = year_counts['photo and video count']

# Create the bar plot
plt.bar(x, height)
plt.xlabel('year')
plt.ylabel('photo and video count')
plt.title('Count of Posted photo and video per year')
plt.xticks(rotation=45, ha='right')
plt.show()


```
